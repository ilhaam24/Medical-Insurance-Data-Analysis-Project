```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\wwwil\Downloads\insurance.csv")
```

```python
df
```

```python
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")
```

```python
df = df.drop_duplicates().reset_index(drop=True)
print(f"Dataset shape after dropping duplicates: {df.shape}")
```

```python
BMI_as_bins = [0, 24.99, 29.99, float('inf')]
BMI_labels = ['Normal', 'Overweight', 'Obese']
```

```python
df['bmi_category'] = pd.cut(df['bmi'], bins=BMI_as_bins, labels=BMI_labels)
```

```python
df.to_csv(r'C:\Users\wwwil\Downloads\Cleaned_insurance_data.csv', index=False)
print("\nSuccess!")
```
